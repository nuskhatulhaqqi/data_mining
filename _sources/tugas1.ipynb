Change Values to 1/0

CertificationCourse:

If values is Yes change to 1

If values is No change to 0

Gender:

If values is Male change to 1

If values is Female change to 0

```python
# embark code
code_CertificationCourse_for_cherbourg = "No"
code_CertificationCourse_for_southampton = "Yes"
code_Gender_for_cherbourg = "Female"
code_Gender_for_southampton = "Male"
# binary value
value_of_one = 1
value_of_zero = 0

def change_code_CertificationCourse_to_biner(CertificationCourse):
    return value_of_one if CertificationCourse == code_CertificationCourse_for_southampton else value_of_zero

def change_code_Gender_to_biner(Gender):
    return value_of_one if Gender == code_Gender_for_southampton else value_of_zero
```

```python
# Update all values of 'Embarked' series
dp["CertificationCourse"] = dp["CertificationCourse"].apply(change_code_CertificationCourse_to_biner)

dp["Gender"] = dp["Gender"].apply(change_code_Gender_to_biner)
```

```python
# CONSTAN VARIABLE
DECREMENT_BY_ONE = 1
INCREMENT_BY_ONE = 1

CONTINGENCY_TABLE_VALUE = {
    "q" : (1,1),
    "r" : (1,0),
    "s" : (0,1),
    "t" : (0,0),
}
```

```python
def get_series(dp, idx, series):
    return dp.loc[(idx - DECREMENT_BY_ONE), series]
```

```python
def get_dissimilarity_dataset(dp, series_index = [], series = []):
    first_series = get_series(dp, series_index[0], series)
    second_series = get_series(dp, series_index[1], series)
    dataset = pd.concat([first_series,second_series],axis=1)
    return dataset.T
```

```python
def count_contingency_value(dp, start_index = 0, last_index = 1):

    CONTINGENCY_VALUE = {
        "q" : 0,
        "r" : 0,
        "s" : 0,
        "t" : 0,
    }

    column_range = dp.shape[1]

    for column in range(column_range):
        for value in CONTINGENCY_TABLE_VALUE:
            item = list((tuple(dp.loc[(start_index - DECREMENT_BY_ONE):(last_index - DECREMENT_BY_ONE), dp.columns[column]]) == CONTINGENCY_TABLE_VALUE[value], value))
            if item[0] == True:
                if item[1] == "q":
                    CONTINGENCY_VALUE["q"] += 1
                if item[1] == "r":
                    CONTINGENCY_VALUE["r"] += 1
                if item[1] == "s":
                    CONTINGENCY_VALUE["s"] += 1
                if item[1] == "t":
                    CONTINGENCY_VALUE["t"] += 1

    return CONTINGENCY_VALUE
```

```python
# d(1,2)
d_1_2 = get_dissimilarity_dataset(dp, [1,2], ["CertificationCourse", "Gender"])
d_1_2
```
